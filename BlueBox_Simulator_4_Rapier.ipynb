# 4.   Search data

**4.1   Search for a (sub)string in the CBSIM IO data frame**

```python
row_search_io = input("Enter search item: ")
search_result_io = df_cbsim_io[df_cbsim_io.apply(lambda row: row.astype(str).str.contains(row_search_io, case=False).any(), axis=1)]
if search_result_io.empty:
  print(Fore.RED + "No matches found" + Fore.RESET)
else:
  print(Fore.GREEN + "Found something" + Fore.RESET)
search_result_io
```

**4.2   Search for a (sub)string in the CBSIM SIGNALS data frame**

```python
row_search_signals = input("Enter search item: ")
search_result_signals = df_cbsim_signals[df_cbsim_signals.apply(lambda row: row.astype(str).str.contains(row_search_signals, case=False).any(), axis=1)]
if search_result_signals.empty:
  print(Fore.RED + "No matches found" + Fore.RESET)
else:
  print(Fore.GREEN + "Found something" + Fore.RESET)
search_result_signals
```

# 5.   Adjust data

Look for a row index in the cbsim io data frame.

When a row is found, the matches in de cbsim signals data frame will be listed.

**5.1   Select row by index of the CBSIM IO dataframe**

```python
row_select = input("Select a row by index: ")
row_index = int(row_select)
if row_index in df_cbsim_io.index:
  print(Fore.GREEN + "- Index " + row_select + " found in df_cbsim_io" + Fore.RESET)
else:
  print(Fore.RED + "- Index " + row_select + " not found in df_cbsim_io - First row is selected" + Fore.RESET)
  row_index = 0
# ------------------------------------------------------------------
search_id = df_cbsim_io.loc[row_index]["SEARCH_ID"]
try:
  id_index = df_cbsim_signals.loc[df_cbsim_signals['ID'] == int(search_id)].index[0]
  print(Fore.GREEN + "- Found ID " + str(df_cbsim_signals.loc[id_index]["ID"]) + " with index " + str(id_index) + " in df_cbsim_signals")
  print(Fore.RESET + df_cbsim_signals.loc[id_index]["CBSIM"] + " - " + df_cbsim_signals.loc[id_index]["CBSIM CIRCUIT"] + " - " + df_cbsim_signals.loc[id_index]["SIGNAL"])
  print('*' * 50)
except Exception as ex:
  print(Fore.RED + "ID not found - Default is loaded")
  print(Fore.RESET + "Exception: ", str(ex))
  id_index = 0
print(Fore.BLUE + "df_cbsim_io" + Fore.RESET)
display(df_cbsim_io.loc[[row_index]])
print(Fore.BLUE + "df_cbsim_signals" + Fore.RESET)
display(df_cbsim_signals.loc[id_index-3:id_index+3])
```

**5.2   Adjust the CBSIM IO data frame**

```python
print("Adjust values for " + df_cbsim_io.loc[row_index]["CBSIM"] + " - " + df_cbsim_io.loc[row_index]["CBSIM_SIGNAL"] + " - " + df_cbsim_io.loc[row_index]["IO_NAME"])
print("Current IO name: " + df_cbsim_io.loc[row_index]["IO_NAME"])
io_name = input("Enter new IO NAME: ")
if not io_name: io_name = df_cbsim_io.loc[row_index]["IO_NAME"]
print("Current comment: " + df_cbsim_io.loc[row_index]["COMMENT"])
comment = input("Enter new COMMENT: ")
if not comment: comment = df_cbsim_io.loc[row_index]["COMMENT"]
df_cbsim_io.at[row_index, "IO_NAME"] = io_name
df_cbsim_io.at[row_index, "COMMENT"] = comment
print(Fore.BLUE + "df_cbsim_io updated" + Fore.RESET)
df_cbsim_io.to_csv(fn_cbsim_io, index=False, sep=';')
display(df_cbsim_io.loc[[row_index]])
```

**5.3   Adjust the CBSIM SIGNALS data frame**

```python
id_index = df_cbsim_signals.loc[df_cbsim_signals['ID'] == int(search_id)].index[0]
print("Adjust values for " + df_cbsim_signals.loc[id_index]["CBSIM"] + " - " + df_cbsim_signals.loc[id_index]["CBSIM CIRCUIT"] + " - " + df_cbsim_signals.loc[id_index]["SIGNAL"])
print("Current SIGNAL: " + df_cbsim_signals.loc[id_index]["SIGNAL"])
signal = input("Enter new SIGNAL: ")
if not signal: signal = df_cbsim_signals.loc[id_index]["SIGNAL"]
print("Current BOARD: " + df_cbsim_signals.loc[id_index]["BOARD"])
board = input("Enter new BOARD: ")
if not board: board = df_cbsim_signals.loc[id_index]["BOARD"]
print("Current CONNECTION ID: " + str(df_cbsim_signals.loc[id_index]["CONNECTION_ID"]))
connection_id = input("Enter new CONNECTION ID: ")
if not connection_id: connection_id =  df_cbsim_signals.loc[id_index]["CONNECTION_ID"]
df_cbsim_signals.at[id_index, "SIGNAL"] = signal
df_cbsim_signals.at[id_index, "BOARD"] = board
df_cbsim_signals.at[id_index, "CONNECTION_ID"] = connection_id
print(Fore.BLUE + "df_cbsim_signals updated" + Fore.RESET)
df_cbsim_signals.to_csv(fn_cbsim_signals, index=False, sep=';')
display(df_cbsim_signals.loc[id_index-3:id_index+3])
```

**5.4   Adjust the CBSIM SIGNALS data frame by a new index**

```python
temp_index = input("Select a row by index: ")
id_index = int(temp_index)
print("Adjust values for " + df_cbsim_signals.loc[id_index]["CBSIM"] + " - " + df_cbsim_signals.loc[id_index]["CBSIM CIRCUIT"] + " - " + df_cbsim_signals.loc[id_index]["SIGNAL"])
print("Current SIGNAL: " + df_cbsim_signals.loc[id_index]["SIGNAL"])
signal = input("Enter new SIGNAL: ")
if not signal: signal = df_cbsim_signals.loc[id_index]["SIGNAL"]
print("Current BOARD: " + df_cbsim_signals.loc[id_index]["BOARD"])
board = input("Enter new BOARD: ")
if not board: board = df_cbsim_signals.loc[id_index]["BOARD"]
print("Current CONNECTION ID: " + str(df_cbsim_signals.loc[id_index]["CONNECTION_ID"]))
connection_id = input("Enter new CONNECTION ID: ")
if not connection_id: connection_id =  df_cbsim_signals.loc[id_index]["CONNECTION_ID"]
df_cbsim_signals.at[id_index, "SIGNAL"] = signal
df_cbsim_signals.at[id_index, "BOARD"] = board
df_cbsim_signals.at[id_index, "CONNECTION_ID"] = connection_id
print(Fore.BLUE + "df_cbsim_signals updated" + Fore.RESET)
df_cbsim_signals.to_csv(fn_cbsim_signals, index=False, sep=';')
display(df_cbsim_signals.loc[id_index-3:id_index+3])
```
